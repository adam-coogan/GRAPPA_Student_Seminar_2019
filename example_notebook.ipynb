### Smallest heading

Generate a bunch of $x$ points, then take

$$y = 3 e^{x}.$$

[Reference 1](http://arxiv.org/)

```python
xs = np.linspace(0, 5, 300)
ys = 3 * np.exp(xs)
```

```python
plt.plot(xs, ys)
plt.xlabel(r"$x$")
plt.ylabel(r"$y(x)$")
```
