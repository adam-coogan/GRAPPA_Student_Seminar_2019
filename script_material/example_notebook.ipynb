# Slicing/list comprehensions

```python
arr = ["a", "b", "c", "d", "e", "f"]
```

```python
print(arr[2])  # python uses 0-based indexing
print(arr[:4])  # slice from beginning to 3rd element
print(arr[2:])  # slice from 2nd element to end
print(arr[3:5])  # slice from 3rd element to 4th
```

# Vectorization in `numpy`

`numpy` functions can generally operator on arrays as if they were just numbers. This lets you write very concise, efficient code, which avoids `for` loops.

```python
xs = np.linspace(0.1, 20, 50)
ys = np.exp((xs**2 - 2 * xs - 1))

plt.loglog(xs, ys, '.')
```

```python
xs = np.logspace(np.log10(0.1), np.log10(20), 50)
ys = np.exp((xs**2 - 2 * xs - 1))

plt.loglog(xs, ys, '.')
```

```python
xs = np.geomspace(0.1, 20, 50)
ys = np.exp((xs**2 - 2 * xs - 1))

plt.loglog(xs, ys, '.')
```

# Root finding example

```python
def f_1(a, x):
    return a * x

def f_2(b, x):
    return b * x**2

def solver(a, b):
    # Note that python is very flexible -- we can define functions inside
    # of other functions. Since function define scope in python, `g` will
    # only be accessible inside `solver`.
    def g(x):  # function composition
        return f_1(a, x) - f_2(a, x) + 10
    
    # `root_scalar` returns an object indicating whether the solver
    # converged, etc.
    return root_scalar(g, x0=0, bracket=[0, 10])

print(solver(3, 2), "\n")

# You can label arguments explicitly, which lets you change their order
print(solver(b=2.5, a=1.5))
```
